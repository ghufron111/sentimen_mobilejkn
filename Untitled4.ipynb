```python
# Necessary imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the dataset
data = pd.read_csv('dataset/mobile_jkn.csv')  # Adjust the file path as necessary

# Convert the 'score' column into binary classes: positive (4 or 5) and negative (1, 2, 3)
data['sentiment'] = data['score'].apply(lambda x: 'positif' if x >= 4 else 'negatif')

# Define features (X) and the binary target variable (y)
X = data['content']  # Assuming 'content' column contains the review text
y = data['sentiment']

# Vectorize the text data using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Limiting to 1000 features for simplicity
X_tfidf = tfidf_vectorizer.fit_transform(X).toarray()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)

# Initialize and train the Naive Bayes classifier
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = nb_classifier.predict(X_test)

# Generate the classification report and accuracy
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=['negatif', 'positif'])

# Display accuracy and classification report
print("Akurasi model tanpa SMOTE:", accuracy)
print("Laporan Klasifikasi:")
print(report)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from keras.models import Sequential
from keras.layers import LSTM, Dense, Embedding, SpatialDropout1D
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
from keras.preprocessing.text import Tokenizer

# Load your dataset
data = pd.read_csv('dataset/mobile_jkn.csv')  # Modify with your actual file path
X = data['content']  # Replace with the appropriate column for your features
y = data['score']  # Replace with your target column

# Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_binary = np.where(y_encoded > 2, 1, 0)  # Group ratings into 2 classes: 0 (negatif) and 1 (positif)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

# Preprocess text data (tokenization, padding)
max_words = 5000  # Number of words to consider as features
max_len = 100  # Maximum length of sequences

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train)

# Convert texts to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad the sequences
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)

# Apply SMOTE to balance the training set based on the encoded target labels
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_pad, y_train)

# Convert labels to categorical
y_train_cat = to_categorical(y_train_res)
y_test_cat = to_categorical(y_test)

# Build the LSTM model
model = Sequential()
model.add(Embedding(max_words, 128, input_length=max_len))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(2, activation='softmax'))  # 2 output classes

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train_res, y_train_cat, epochs=5, batch_size=64, validation_data=(X_test_pad, y_test_cat))

# Evaluate the model
loss, accuracy = model.evaluate(X_test_pad, y_test_cat, verbose=0)
print(f"Model Accuracy: {accuracy:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from imblearn.over_sampling import SMOTE
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
```

```python
# Load dataset
df = pd.read_csv('dataset/mobile_jkn.csv')

# Data preprocessing
X = df['content']  # Assuming 'review' column has the text
y = df['score']  # Assuming 'rating' column has the labels

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Preprocessing text data
max_words = 10000
max_len = 100

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)

# Function to evaluate models
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    
    return accuracy, precision, recall, f1

# Results dictionary
results = {
    'Model': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': []
}
```

```python
# LSTM without SMOTE
model_without_smote = Sequential()
model_without_smote.add(Embedding(max_words, 128, input_length=max_len))
model_without_smote.add(LSTM(100))
model_without_smote.add(Dense(2, activation='softmax'))
model_without_smote.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_without_smote.fit(X_train_pad, y_train, epochs=5, batch_size=64, verbose=0)

y_pred_lstm_without_smote = np.argmax(model_without_smote.predict(X_test_pad), axis=1)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_lstm_without_smote)

results['Model'].append('LSTM (Without SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)

# LSTM with SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE on the training data before padding
X_train_res, y_train_res = smote.fit_resample(X_train_pad, y_train)

model_with_smote = Sequential()
model_with_smote.add(Embedding(max_words, 128, input_length=max_len))
model_with_smote.add(LSTM(100))
model_with_smote.add(Dense(2, activation='softmax'))
model_with_smote.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_with_smote.fit(X_train_res, y_train_res, epochs=5, batch_size=64, verbose=0)

y_pred_lstm_with_smote = np.argmax(model_with_smote.predict(X_test_pad), axis=1)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_lstm_with_smote)

results['Model'].append('LSTM (With SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)
```

```python
# Naive Bayes without SMOTE
nb_model_without_smote = MultinomialNB()
nb_model_without_smote.fit(X_train_pad, y_train)

y_pred_nb_without_smote = nb_model_without_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_nb_without_smote)

results['Model'].append('Naive Bayes (Without SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)

# Naive Bayes with SMOTE
nb_model_with_smote = MultinomialNB()
nb_model_with_smote.fit(X_train_res, y_train_res)

y_pred_nb_with_smote = nb_model_with_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_nb_with_smote)

results['Model'].append('Naive Bayes (With SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)
```

```python
# Random Forest without SMOTE
rf_model_without_smote = RandomForestClassifier()
rf_model_without_smote.fit(X_train_pad, y_train)

y_pred_rf_without_smote = rf_model_without_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_rf_without_smote)

results['Model'].append('Random Forest (Without SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)

# Random Forest with SMOTE
rf_model_with_smote = RandomForestClassifier()
rf_model_with_smote.fit(X_train_res, y_train_res)

y_pred_rf_with_smote = rf_model_with_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_rf_with_smote)

results['Model'].append('Random Forest (With SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)
```

```python
# SVM without SMOTE
svm_model_without_smote = SVC(probability=True)
svm_model_without_smote.fit(X_train_pad, y_train)

y_pred_svm_without_smote = svm_model_without_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_svm_without_smote)

results['Model'].append('SVM (Without SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)

# SVM with SMOTE
svm_model_with_smote = SVC(probability=True)
svm_model_with_smote.fit(X_train_res, y_train_res)

y_pred_svm_with_smote = svm_model_with_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_svm_with_smote)

results['Model'].append('SVM (With SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)
```

```python
# XGBoost without SMOTE
xgb_model_without_smote = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_model_without_smote.fit(X_train_pad, y_train)

y_pred_xgb_without_smote = xgb_model_without_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_xgb_without_smote)

results['Model'].append('XGBoost (Without SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)

# XGBoost with SMOTE
xgb_model_with_smote = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_model_with_smote.fit(X_train_res, y_train_res)

y_pred_xgb_with_smote = xgb_model_with_smote.predict(X_test_pad)
accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred_xgb_with_smote)

results['Model'].append('XGBoost (With SMOTE)')
results['Accuracy'].append(accuracy)
results['Precision'].append(precision)
results['Recall'].append(recall)
results['F1 Score'].append(f1)
```

```python
# Create a DataFrame for results
results_df = pd.DataFrame(results)

# Display results
print(results_df)
```
